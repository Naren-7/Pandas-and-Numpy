## Importando **Pandas and NumPay**

```python
import pandas as pd
import numpy as np
```

```python
df_books = pd.read_csv(r'./Dates/bestsellers-with-categories.csv', sep=',', header=0)
df_books.head(2)
```

## **Groupby:** Agrupamiento

```python
df_books.groupby(by='Author').count()
```

```python
df_books.groupby(by='Author').mean()
```

```python
df_books.groupby(by="Author").describe()
```

```python
df_books.groupby(by='Author').max().loc['William Davis']
```

## Quitamos el indice de la agrupacion 

```python
df_books.groupby('Author').sum().reset_index()
```

## agrupacion por **agregacion** 

```python
df_books.groupby('Author').agg(['min', 'max'])
```

```python
df_books.head(2)
```

```python
df_books.groupby('Author').agg({ 'Reviews': [ 'min', 'max'], 'User Rating': 'sum' })
```

```python
df_books.groupby(by="Author").agg({'Reviews': ['max', 'min'], 'Price': ['sum']})
```

```python
df_books.groupby(['Author', 'Year']).count()
```

## Extra

```python
df_books.groupby("Author",sort=True).apply(func="min")
```
