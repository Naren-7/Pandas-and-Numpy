```python
import pandas as pd
import numpy as np
```

```python
df_books = pd.read_csv(r'./Dates/bestsellers-with-categories.csv', sep=',')
```

```python
df_books.head()
```

## Pivot Table
- tabla dinamicas que nos permiten modificar nuestra data frame de la forma que mas interese. 

```python
df_books.pivot_table(index='Author',columns='Genre',values='User Rating', dropna=True, )
```

```python
df_books.pivot_table(index='Genre',columns='Year', values='Price',aggfunc='sum',fill_value=1, observed=True, margins_name="H" ) 
```

```python
df_books.pivot_table(index='Author', columns="Year", values="Reviews", aggfunc="sum", fill_value="" ).sort_values(by="Author",ascending=True,)
```

## Melt(derretir)

- Transforma **columnas** en **filas**

```python
df_books[['Name','Genre']].head(5)
```

```python
df_books[['Name','Genre']].head(5).melt()
```

```python
df_books.melt(id_vars='Year',value_vars='Genre', ignore_index=True)
```

```python
df_books.melt(id_vars=('Name', 'Year'), var_name='Name', )
```
