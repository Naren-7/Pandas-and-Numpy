# **Loc**

- **Loc** esta basado en **Label(etiquetas del dataframe)** la **'L'** lo referencia.

```python
df_books.loc[:]
```

```python
df_books.loc[0:4, ['Nameee', 'Authorrr', 'Year'] ]
```

```python
df_books.loc[ :, ['Reviews'] ] * -3
```

```python
df_books.loc[ :, ['Authorrr'] ] == 'JJ Smith'
```

## **Iloc**

- Es muy parecido a **Loc**
- Busca por **'I'**  **Indice** no por Label 

```python
df_books.iloc[:]
```

```python
df_books.iloc[:, 0:3]
```

```python
df_books.iloc[:, 1] = 'Nada' 
df_books
```

```python
df_books.iloc[:2, 2:]
```

```python
df_books.iloc[1,3] / 3
```

```python
df_books.iloc[:3,3:]
```
